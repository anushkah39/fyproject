```python
import cv2

def preprocess_image(input_path, output_path, target_width=1024):
    """
    Preprocess the input image: resize, grayscale, blur, threshold, and sharpen.
    Args:
        input_path (str): Path to the input image.
        output_path (str): Path to save the preprocessed image.
        target_width (int): Target width for resizing while maintaining aspect ratio.
    """
    # Load the original image
    image = cv2.imread(input_path)

    # Resize the image while maintaining aspect ratio
    aspect_ratio = image.shape[0] / image.shape[1]
    new_dimensions = (target_width, int(target_width * aspect_ratio))
    resized_image = cv2.resize(image, new_dimensions, interpolation=cv2.INTER_AREA)

    # Convert to grayscale
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian Blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (3, 3), 0)

    # Apply adaptive thresholding for binarization
    thresholded = cv2.adaptiveThreshold(
        blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )

    # Sharpen the image
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 1))
    sharpened = cv2.morphologyEx(thresholded, cv2.MORPH_CLOSE, kernel)

    # Save the preprocessed image
    cv2.imwrite(output_path, sharpened)
    print(f"Preprocessed and saved: {output_path}")
```

```python
def preprocess_all_images(input_dir, output_dir):
    """
    Preprocess all images in the input directory and save them to the output directory.
    Args:
        input_dir (str): Path to the directory containing input images.
        output_dir (str): Path to save preprocessed images.
    """
    for root, _, files in os.walk(input_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter image files
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, input_dir)
                output_path = os.path.join(output_dir, relative_path)

                # Ensure subdirectories exist in the output directory
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Preprocess the image
                preprocess_image(input_path, output_path)

# Preprocess all images
preprocess_all_images(input_dir, preprocessed_dir)
```

```python
import requests
import json

# Load Azure configuration
azure_config_path = '/content/drive/My Drive/Handwritten_paper_evaluation/azure_config.json'
with open(azure_config_path, 'r') as f:
    azure_config = json.load(f)

AZURE_SUBSCRIPTION_KEY = azure_config['AZURE_SUBSCRIPTION_KEY']
AZURE_ENDPOINT = azure_config['AZURE_ENDPOINT']

def perform_ocr(image_path):
    """
    Perform OCR on the given image using Azure Computer Vision.
    Args:
        image_path (str): Path to the input image.
    Returns:
        dict: OCR results in JSON format.
    """
    with open(image_path, 'rb') as image_file:
        image_data = image_file.read()

    # Define the OCR endpoint and headers
    ocr_url = f"{AZURE_ENDPOINT}/vision/v3.2/read/analyze"
    headers = {
        'Ocp-Apim-Subscription-Key': AZURE_SUBSCRIPTION_KEY,
        'Content-Type': 'application/octet-stream'
    }

    # Send the image to Azure OCR
    response = requests.post(ocr_url, headers=headers, data=image_data)

    # Check if the request was successful
    if response.status_code == 202:
        operation_url = response.headers['Operation-Location']
        return operation_url
    else:
        raise Exception(f"OCR request failed with status code {response.status_code}: {response.json()}")

def get_ocr_results(operation_url):
    """
    Retrieve OCR results from Azure Computer Vision.
    Args:
        operation_url (str): Operation URL for retrieving OCR results.
    Returns:
        dict: OCR results in JSON format.
    """
    headers = {'Ocp-Apim-Subscription-Key': AZURE_SUBSCRIPTION_KEY}
    while True:
        response = requests.get(operation_url, headers=headers)
        if response.status_code == 200:
            result = response.json()
            if result['status'] == 'succeeded':
                return result['analyzeResult']
            elif result['status'] == 'failed':
                raise Exception("OCR processing failed.")
        elif response.status_code == 429:
            print("Rate limit exceeded. Retrying...")
        else:
            raise Exception(f"Error retrieving OCR results: {response.status_code}")
```

```python
def ocr_all_images(preprocessed_dir, ocr_results_dir):
    """
    Perform OCR on all preprocessed images and save results to a directory.
    Args:
        preprocessed_dir (str): Directory containing preprocessed images.
        ocr_results_dir (str): Directory to save OCR results.
    """
    for root, _, files in os.walk(preprocessed_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, preprocessed_dir)
                output_path = os.path.join(ocr_results_dir, relative_path.replace('.jpg', '.json'))

                # Ensure subdirectories exist in the output directory
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Perform OCR
                try:
                    operation_url = perform_ocr(input_path)
                    ocr_results = get_ocr_results(operation_url)

                    # Save OCR results to JSON
                    with open(output_path, 'w') as json_file:
                        json.dump(ocr_results, json_file, indent=4)
                    print(f"OCR results saved: {output_path}")
                except Exception as e:
                    print(f"Error processing {input_path}: {e}")

# Perform OCR on all preprocessed images
ocr_all_images(preprocessed_dir, ocr_results_dir)
```

```python
def reconstruct_and_save_text(ocr_file_path, output_path):
    """
    Reconstruct plain text from OCR JSON and save it to a file.
    Args:
        ocr_file_path (str): Path to the OCR JSON file.
        output_path (str): Path to save the reconstructed text file.
    """
    # Load OCR JSON
    with open(ocr_file_path, 'r') as json_file:
        ocr_result = json.load(json_file)

    # Reconstruct plain text
    reconstructed_text = []
    for page in ocr_result.get('readResults', []):
        reconstructed_text.append(f"Page {page['page']}:")
        for line in page.get('lines', []):
            reconstructed_text.append(line['text'])
    plain_text = "\n".join(reconstructed_text)

    # Save plain text to file
    with open(output_path, 'w') as text_file:
        text_file.write(plain_text)
    print(f"Reconstructed text saved at: {output_path}")
```

```python
def process_all_ocr_results(ocr_results_dir, reconstructed_text_dir):
    """
    Process all OCR JSON files, reconstruct text, and save as plain text files.
    Args:
        ocr_results_dir (str): Directory containing OCR JSON files.
        reconstructed_text_dir (str): Directory to save reconstructed text files.
    """
    for root, _, files in os.walk(ocr_results_dir):
        for file in files:
            if file.lower().endswith('.json'):  # Process only JSON files
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, ocr_results_dir)
                output_path = os.path.join(reconstructed_text_dir, relative_path.replace('.json', '.txt'))

                # Ensure subdirectories exist in the reconstructed text directory
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Reconstruct and save text
                reconstruct_and_save_text(input_path, output_path)

# Process all OCR results
process_all_ocr_results(ocr_results_dir, reconstructed_text_dir)
```

```python
import re
import json
import os

def segment_text(input_text):
    """
    Segment text into categories: header, printed text, theory, and code.

    Args:
        input_text (str): The text to be segmented.

    Returns:
        dict: Segmented text in categories.
    """
    segments = {
        "header": "",
        "printed_text": "",
        "theory_answers": "",
        "code_answers": []
    }

    # Patterns to identify printed text
    printed_text_patterns = [
        r"\bKRISH\b", r"\bTeacher's Signature\b", r"\bAMAR\b",
        r"\bDate\b", r"\bPage\b"
    ]

    lines = input_text.split("\n")
    current_section = None

    for line in lines:
        line = line.strip()

        # Skip empty lines
        if not line:
            continue

        # Identify printed text
        if any(re.search(pattern, line, re.IGNORECASE) for pattern in printed_text_patterns):
            if current_section != "printed_text":
                current_section = "printed_text"
            segments["printed_text"] += f"{line} "

        # Identify header lines (e.g., metadata like name, date)
        elif re.match(r"^(Name|Roll|Class|Section|Branch|Page|Date)", line, re.IGNORECASE):
            if current_section != "header":
                current_section = "header"
            segments["header"] += f"{line} "

        # Identify code blocks (contains code syntax patterns)
        elif re.search(r"(def |if |else|while |for |print|#|=|:|%|\bint\b|\bfloat\b|\bchar\b)", line):
            if current_section != "code_answers":
                current_section = "code_answers"
            segments["code_answers"].append(line)

        # Everything else is assumed to be theoretical answers
        else:
            if current_section != "theory_answers":
                current_section = "theory_answers"
            segments["theory_answers"] += f"{line} "

    # Format code answers as a single block
    segments["code_answers"] = "\n".join(segments["code_answers"])

    return segments

def process_and_segment_all_texts(input_dir, output_dir):
    """
    Process all text files in the input directory, segment them, and save the results as JSON.

    Args:
        input_dir (str): Directory containing text files.
        output_dir (str): Directory to save segmented JSON files.
    """
    os.makedirs(output_dir, exist_ok=True)

    for root, _, files in os.walk(input_dir):
        for file in files:
            if file.lower().endswith('.txt'):  # Process only text files
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, input_dir)
                output_path = os.path.join(output_dir, relative_path.replace('.txt', '_segmented.json'))

                # Load the input text
                with open(input_path, 'r') as text_file:
                    input_text = text_file.read()

                # Segment the text
                segmented_output = segment_text(input_text)

                # Save segmented output to JSON
                os.makedirs(os.path.dirname(output_path), exist_ok=True)
                with open(output_path, 'w') as json_file:
                    json.dump(segmented_output, json_file, indent=4)

                print(f"Segmented text saved to {output_path}")

# Define input and output directories
reconstructed_text_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Reconstructed_Text/'
segmented_output_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Segmented_Text/'

# Process all text files and segment them
process_and_segment_all_texts(reconstructed_text_dir, segmented_output_dir)
```

```python
import os
import json

def merge_segmented_files(input_dir, output_file):
    """
    Merge all segmented JSON files into a single readable text file for inspection.

    Args:
        input_dir (str): Directory containing segmented JSON files.
        output_file (str): Path to the merged output file.
    """
    with open(output_file, 'w') as outfile:
        for root, _, files in os.walk(input_dir):
            for file in sorted(files):  # Sort files for consistent order
                if file.lower().endswith('_segmented.json'):
                    file_path = os.path.join(root, file)

                    # Load segmented JSON
                    with open(file_path, 'r') as json_file:
                        segmented_data = json.load(json_file)

                    # Write file header
                    outfile.write(f"--- {file} ---\n\n")

                    # Write each section in a readable format
                    outfile.write("Header:\n")
                    outfile.write(segmented_data.get("header", "") + "\n\n")

                    outfile.write("Printed Text:\n")
                    outfile.write(segmented_data.get("printed_text", "") + "\n\n")

                    outfile.write("Theoretical Answers:\n")
                    outfile.write(segmented_data.get("theory_answers", "") + "\n\n")

                    outfile.write("Code Answers:\n")
                    outfile.write(segmented_data.get("code_answers", "") + "\n\n")

                    outfile.write("\n" + "="*50 + "\n\n")  # Separator between files

    print(f"Segmented data merged into: {output_file}")

# Define the input directory and output file path
segmented_text_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Segmented_Text/'
merged_segmented_file_path = '/content/drive/My Drive/Handwritten_paper_evaluation/Merged_Segmented_Text.txt'

# Merge all segmented JSON files into a single readable file
merge_segmented_files(segmented_text_dir, merged_segmented_file_path)
```

```python
import re

def refine_segmented_data(input_file, output_file):
    """
    Refine the segmented data for better accuracy in headers, printed text,
    theoretical answers, and code answers.

    Args:
        input_file (str): Path to the input JSON file.
        output_file (str): Path to save the refined JSON file.
    """
    with open(input_file, 'r') as infile:
        data = json.load(infile)

    refined_data = {}

    # Refine Header
    raw_header = data.get("header", "")
    refined_data["header"] = re.sub(r"[\s:]+", " ", raw_header).strip()

    # Refine Printed Text
    raw_printed_text = data.get("printed_text", "")
    refined_data["printed_text"] = re.sub(
        r"(Page\s*\d+|Date\s*[:\-]*|Teacher's Signature|KRISH|AMAR)", "", raw_printed_text
    ).strip()

    # Refine Theoretical Answers
    raw_theory = data.get("theory_answers", "")
    refined_data["theory_answers"] = re.sub(
        r"[^\w\s,.()\"'-]|(\s*\n\s*)+", " ", raw_theory
    ).strip()

    # Refine Code Answers
    raw_code = data.get("code_answers", "")
    refined_data["code_answers"] = re.sub(
        r"(Page\s*\d+|Teacher's Signature|AMAR|KRISH)|(\s*\n\s*)+", " ", raw_code
    ).strip()

    # Save refined data
    with open(output_file, 'w') as outfile:
        json.dump(refined_data, outfile, indent=4)
    print(f"Refined segmented data saved: {output_file}")
```

```python
def process_refinement(segmented_dir, refined_dir):
    """
    Refine all segmented JSON files in the directory.

    Args:
        segmented_dir (str): Path to the directory containing segmented JSON files.
        refined_dir (str): Path to save the refined JSON files.
    """
    for root, _, files in os.walk(segmented_dir):
        for file in files:
            if file.lower().endswith('_segmented.json'):
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, segmented_dir)
                output_path = os.path.join(refined_dir, relative_path)

                # Ensure output directory exists
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Refine the segmented JSON
                refine_segmented_data(input_path, output_path)

# Define paths
segmented_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Segmented_Text/'
refined_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Refined_Text/'

# Process all files
process_refinement(segmented_dir, refined_dir)
```

```python
import os
import re
import json

# Paths
ocr_results_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/OCR_Results/'
refined_segmented_text_dir = '/content/drive/My Drive/Handwritten_paper_evaluation/Refined_Segmented_Text/'
os.makedirs(refined_segmented_text_dir, exist_ok=True)

def segment_text_v2(ocr_result):
    """
    Enhanced segmentation logic for OCR text using heuristics.
    Args:
        ocr_result (dict): Parsed OCR JSON data.
    Returns:
        dict: Segmented text with headers, printed text, theoretical answers, and code answers.
    """
    header, printed_text, theory, code = [], [], [], []

    # Loop through all recognized lines
    for page in ocr_result.get("readResults", []):
        for line in page.get("lines", []):
            text = line['text'].strip()

            # Header Detection Heuristic
            if re.search(r'^(Name|Roll|Class|Section|Date|Page)', text, re.IGNORECASE):
                header.append(text)
            # Printed Text Detection
            elif re.search(r'(Date|Teacher|Page|Signature|KRISH|AMAR)', text, re.IGNORECASE):
                printed_text.append(text)
            # Code Detection Heuristic
            elif re.search(r'(\bint\b|\bfloat\b|\bif\b|{|}|:|;|print|\breturn\b|def|class)', text):
                code.append(text)
            # Theoretical Answer
            else:
                theory.append(text)

    return {
        "Header": " ".join(header),
        "Printed Text": " ".join(printed_text),
        "Theoretical Answers": " ".join(theory),
        "Code Answers": "\n".join(code)
    }

def process_all_ocr_results_v2(ocr_results_dir, refined_segmented_text_dir):
    """
    Process all OCR results and create segmented JSON files for each.
    Args:
        ocr_results_dir (str): Directory containing OCR JSON files.
        refined_segmented_text_dir (str): Directory to save segmented text files.
    """
    for root, _, files in os.walk(ocr_results_dir):
        for file in files:
            if file.lower().endswith('.json'):  # Process only JSON files
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, ocr_results_dir)
                output_path = os.path.join(refined_segmented_text_dir, relative_path)

                # Ensure subdirectories exist in the segmented text directory
                os.makedirs(os.path.dirname(output_path), exist_ok=True)

                # Read OCR JSON file
                with open(input_path, 'r') as json_file:
                    ocr_result = json.load(json_file)

                # Segment text
                segmented_text = segment_text_v2(ocr_result)

                # Save segmented text to JSON
                output_path = output_path.replace('.json', '_segmented.json')
                with open(output_path, 'w') as json_output_file:
                    json.dump(segmented_text, json_output_file, indent=4)
                print(f"Segmented text saved: {output_path}")

# Process all OCR results
process_all_ocr_results_v2(ocr_results_dir, refined_segmented_text_dir)
```

```python
from difflib import SequenceMatcher

# Extract model answers for Q1
model_theoretical = preprocess_text(model_answers['model_answers'][0]['content']['text'])
model_code = preprocess_text(' '.join(model_answers['model_answers'][1]['content']['code']))

# Function to calculate similarity
def calculate_similarity(text1, text2):
    return SequenceMatcher(None, text1, text2).ratio()

# Calculate similarity for theoretical and code answers
theoretical_similarity = calculate_similarity(preprocessed_theoretical, model_theoretical)
code_similarity = calculate_similarity(preprocessed_code, model_code)

# Display similarity scores
print(f"Theoretical Answer Similarity: {theoretical_similarity * 100:.2f}%")
print(f"Code Answer Similarity: {code_similarity * 100:.2f}%")

# Highlight differences using SequenceMatcher
def highlight_differences(text1, text2):
    matcher = SequenceMatcher(None, text1, text2)
    diff = []
    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag == 'replace':
            diff.append(f"[REPLACE: {text1[i1:i2]} -> {text2[j1:j2]}]")
        elif tag == 'delete':
            diff.append(f"[DELETE: {text1[i1:i2]}]")
        elif tag == 'insert':
            diff.append(f"[INSERT: {text2[j1:j2]}]")
    return ' '.join(diff)

# Display differences
print("\nDifferences in Theoretical Answers:")
print(highlight_differences(preprocessed_theoretical, model_theoretical))

print("\nDifferences in Code Answers:")
print(highlight_differences(preprocessed_code, model_code))
```

```python
import os
import json
from difflib import SequenceMatcher
from pathlib import Path

# Path to the root folder containing student subfolders
root_folder = "/content/drive/MyDrive/Handwritten_paper_evaluation/Refined_Segmented_Text/"
model_answers_file = "/content/drive/MyDrive/Handwritten_paper_evaluation/model_answers.json"
output_file = "/content/drive/MyDrive/Handwritten_paper_evaluation/evaluation_results.json"

# Load model answers
with open(model_answers_file, 'r') as file:
    model_answers = json.load(file)['model_answers']

# Helper function to calculate similarity
def calculate_similarity(text1, text2):
    return round(SequenceMatcher(None, text1, text2).ratio() * 100, 2)

# Evaluation function
def evaluate_answers(student_answers, model_answer, question_type):
    if question_type == "theory":
        similarity = calculate_similarity(student_answers.get("Theoretical Answers", ""), model_answer["content"]["text"])
        feedback = f"Theoretical Answer Similarity: {similarity}%"
        marks = round(similarity * sum(model_answer["marks_allocation"].values()) / 100, 2)
    elif question_type == "code":
        similarity = calculate_similarity(student_answers.get("Code Answers", ""), "\n".join(model_answer["content"]["code"]))
        feedback = f"Code Answer Similarity: {similarity}%"
        marks = round(similarity * model_answer["marks_allocation"] / 100, 2)
    else:
        similarity = 0
        feedback = "Invalid question type."
        marks = 0
    return similarity, feedback, marks

# Iterate through all student subfolders
evaluation_results = []
student_folders = sorted(Path(root_folder).iterdir(), key=os.path.basename)

for student_folder in student_folders:
    if not student_folder.is_dir():
        continue

    student_id = student_folder.name
    student_result = {"student_id": student_id, "q1": {}, "q2": {}}

    for question_file in ["q1_segmented.json", "q2_segmented.json"]:
        question_path = student_folder / question_file
        if not question_path.exists():
            print(f"File {question_path} not found, skipping...")
            continue

        with open(question_path, 'r') as file:
            student_answers = json.load(file)

        # Determine question type
        if question_file.startswith("q1"):
            question_type = "theory"
            model_answer = next(item for item in model_answers if item["question_id"] == "Q1")
        elif question_file.startswith("q2"):
            question_type = "code"
            model_answer = next(item for item in model_answers if item["question_id"] == "Q2")
        else:
            print(f"Unknown question file: {question_file}")
            continue

        # Evaluate the student's answer
        similarity, feedback, marks = evaluate_answers(student_answers, model_answer, question_type)

        # Store the evaluation results
        student_result[question_file.split('_')[0]] = {
            "similarity": similarity,
            "feedback": feedback,
            "marks": marks
        }

    # Append the results to the evaluation list
    evaluation_results.append(student_result)

# Save the evaluation results to a JSON file
with open(output_file, 'w') as file:
    json.dump(evaluation_results, file, indent=4)

print(f"Evaluation completed. Results saved to {output_file}")
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Path to the evaluation results file
evaluation_results_file = '/content/drive/MyDrive/Handwritten_paper_evaluation/evaluation_results.json'

# Load the evaluation results
with open(evaluation_results_file, 'r') as file:
    evaluation_results = json.load(file)

# Initialize lists for storing scores
student_ids = []
theoretical_similarities = []
code_similarities = []
theoretical_marks = []
code_marks = []

# Extract data from the results
for student in evaluation_results:
    student_ids.append(student["student_id"])
    theoretical_similarities.append(student["q1"]["similarity"])
    code_similarities.append(student["q2"]["similarity"])
    theoretical_marks.append(student["q1"]["marks"])
    code_marks.append(student["q2"]["marks"])

# Convert lists to numpy arrays for easier manipulation
theoretical_similarities = np.array(theoretical_similarities)
code_similarities = np.array(code_similarities)
theoretical_marks = np.array(theoretical_marks)
code_marks = np.array(code_marks)

# Calculate averages
avg_theoretical_similarity = np.mean(theoretical_similarities)
avg_code_similarity = np.mean(code_similarities)
avg_theoretical_marks = np.mean(theoretical_marks)
avg_code_marks = np.mean(code_marks)

# Visualization

# Similarity Distribution
plt.figure(figsize=(10, 6))
plt.hist(theoretical_similarities, bins=10, alpha=0.7, label='Theoretical Similarity (%)')
plt.hist(code_similarities, bins=10, alpha=0.7, label='Code Similarity (%)')
plt.title('Similarity Distribution')
plt.xlabel('Similarity (%)')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(True)
plt.show()

# Marks Distribution
plt.figure(figsize=(10, 6))
plt.hist(theoretical_marks, bins=10, alpha=0.7, label='Theoretical Marks')
plt.hist(code_marks, bins=10, alpha=0.7, label='Code Marks')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(True)
plt.show()

# Average Scores
categories = ['Theoretical', 'Code']
average_scores = [avg_theoretical_marks, avg_code_marks]
average_similarities = [avg_theoretical_similarity, avg_code_similarity]

# Average Marks
plt.figure(figsize=(8, 5))
plt.bar(categories, average_scores, color='blue', alpha=0.6)
plt.title('Average Marks')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--')
plt.show()

# Average Similarities
plt.figure(figsize=(8, 5))
plt.bar(categories, average_similarities, color='green', alpha=0.6)
plt.title('Average Similarities')
plt.ylabel('Similarity (%)')
plt.grid(axis='y', linestyle='--')
plt.show()
```
